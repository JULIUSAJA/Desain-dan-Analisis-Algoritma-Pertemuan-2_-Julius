```python
def getSum(myList):
  sum = 0
  for row in myList:
    for item in row:
      sum = sum + item
  return sum

def getBagiSet(setA, setB):
  sumA = getSum(setA)
  sumB = getSum(setB)
  return sumA / sumB

getBagiSet([[1,2,3]], [[3,4,6]])
```

```python
def getSum(myList):
  total = 0
  for baris in myList:
    for item in baris:
      total = total + item
  return total

def getPengurangan(paketA, paketB):
  totalA = getSum(paketA)
  totalB = getSum(paketB)
  return totalA - totalB

getPengurangan([[5,2,1]], [[60,2,1]])
```

```python
def searchBinary(myList, item):
  first = 0
  last = len(myList)-1
  foundFlag = False
  while( first <= last and not foundFlag):
    mid = (first + last)//2
    if myList[mid] == item:
      foundFlag = True
    else:
      if item < myList[mid]:
        last = mid - 1
      else:
        first = mid + 1
  return foundFlag

searchBinary([8,9,10,100,1000,2000,3000], 10)
```
